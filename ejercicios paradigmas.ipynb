## 12. Escribe una función para verificar si una contraseña dada es fuerte (al menos 8 caracteres, al menos una minúscula, al menos una mayúscula, al menos un número y al menos un carácter especial). Debería devolver True si es fuerte y False si no lo es. Se permit el uso de regex

```python
#tu codigo aquí
def is_password_strong_imperative_paradigm(password: str) -> bool:
    password_is_strong = True
    #length check
    if len(password) < 8:
        password_is_strong = False
    #lowercase check
    password_contains_lowercase_letters = False
    for char in password:
        if char.islower():
            password_contains_lowercase_letters = True
    if not password_contains_lowercase_letters:
        password_is_strong = False
    #uppercase check
    password_contains_uppercase_letters = False
    for char in password:
        if char.isupper():
            password_contains_uppercase_letters = True
    if not password_contains_uppercase_letters:
        password_is_strong = False
    #number check
    password_contains_numbers = False
    for char in password:
        if char.isnumeric():
            password_contains_numbers = True
    if not password_contains_numbers:
        password_is_strong = False
    #special character check
    password_contains_special_characters = False
    for char in password:
        if not char.isalnum():
            password_contains_special_characters = True
    if not password_contains_special_characters:
        password_is_strong = False
    
    return password_is_strong

short_password = "p123456"
long_password = "12345678"
no_lowercase_long_password = "P1234567"
lowercase_long_password = "p1234567"
lowercase_uppercase_long_password = "pP234567"
strong_password = "12345Pp."
assert(is_password_strong_imperative_paradigm(strong_password))
```

```python
# Para testear tu función
test_pass(is_password_strong_imperative_paradigm)
```

```python
class PasswordManager:
    def is_password_strong_oop_paradigm(self,given_password):
        return is_password_strong_imperative_paradigm(given_password)
```

```python
password_manager = PasswordManager()
test_pass(password_manager.is_password_strong_oop_paradigm)
```

```python
is_password_strong_functional_paradigm = lambda password: is_password_strong_imperative_paradigm(password)
```
